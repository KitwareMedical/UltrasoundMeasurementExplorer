```python
import numpy as np
from numpy import fft
```

```python
import matplotlib.pyplot as plt
```

```python
tstep = 1. / 30000000
t = np.linspace(-127 * tstep, 128 * tstep, 256)

carrier = np.sin(5000000 * t * 2 * np.pi)
envelope = np.e ** (-(4000000 * t)**2)

plt.plot(carrier * envelope)
plt.show()
```

```python
z = np.fft.fft(carrier * envelope)
```

```python
plt.plot((z * z.conjugate())[:])
plt.show()
```

```python
q = np.fft.fft((z))
plt.plot(q)
plt.show()
```
